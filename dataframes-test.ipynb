```python
import pandas as pd
```

Load the file african_econ_crises.csv

```python
df = pd.read_csv("./african_econ_crisis.csv")
df
```

Read the first 5 lines

```python
df.head(5)
```

How many countries have inflation data in the dataset?

```python
df["country"].count()
```

Which countries are represented in the dataset?

```python
df["country"]
```

Which country has the highest inflation?

```python
df.groupby("country")["inflation_annual_cpi"].max().reset_index().head(1)
```

In what year did Kenya experience its first systemic crisis?

```python
df[df["country"] == "Kenya"].groupby("year").sum().sort_values(by="year",ascending=True).reset_index().head(1)
```

How many systemic crises have occurred in each country?

```python
df.groupby("country")["systemic_crisis"].sum().sort_values(ascending=False).reset_index()
```

How many debt defaults has Zimbabwe had?

```python
df[df["country"] == "Zimbabwe"]["domestic_debt_in_default"].sum()
```

What is the highest exch_usd value for Algeria, and in which year did it occur?

```python
df[df["country"] == "Algeria"].groupby("year").sum().sort_values(by="exch_usd",ascending=False).reset_index().head(1)
```
