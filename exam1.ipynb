```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

Cargamos el conjunto de datos utilizando pd.read_csv, ya que nuestro archivo está en formato CSV. Luego, con df, mostramos los datos en forma de tabla.

```python
df = pd.read_csv('heart.csv')
df
```

Con .head(5) cargamos específicamente las primeras cinco líneas del conjunto de datos, lo que nos permite observar los atributos del archivo y tener una vista preliminar de su estructura.

```python
df.head(5)
```

Con .info() mostramos la información las estadisticas de cada variable

```python
df.info()
```

Para crear un gráfico de la distribución del sexo en el conjunto de datos, utilizamos sns.displot(df["sex"], bins=3), lo que genera un histograma con tres bins lo cual queda mas grande. Luego, con plt.show(), mostramos el gráfico.

```python
sns.displot(df["sex"], bins=3)
plt.show()
```

Para mostrar un histograma de la distribución de la edad en el conjunto de datos, con la cantidad que hay de cada edad, utilizamos sns.displot(df["age"], kde=True).

```python
sns.displot(df["age"], kde=True)
plt.show()
```

Usamos sns.boxplot(data=df, x="cp", y="age") para visualizar la distribución de la edad en función de la variable categórica cp. Este gráfico de caja nos permite identificar la mediana, los cuartiles y posibles valores atípicos en la distribución de la edad para cada categoría de cp.

```python
sns.boxplot(data=df, x="cp", y="age")
plt.show()
```

En este gráfico mostramos la cantidad de personas que tienen frecuencia cardiaca máxima, podemos ver que la mayor cantidad de personas tienen entre 130 y 170.

```python
sns.displot(df["thalach"], kde=True)
plt.show()
```

Este permite comparar cómo varía la edad según el tipo de dolor en el pecho (cp) y la presencia de azúcar en sangre alto (fbs). Se ve que en algunos valores de cp, los pacientes con fbs=1 tienden a ser mayores que aquellos con fbs=0,

```python
sns.boxplot(data=df, x="cp", y="age", hue="fbs")
plt.show()
```

Mostramos los pacientes por su edad que tienen presión arterial en reposo, podemos ver que la mayoría de personas con presión arterial en reposo están entre los 50 y los 70 años

```python
sns.scatterplot(data=df, x="age", y="trestbps")
plt.show()
```

En el histograma observamos la cantidad de personas según sus niveles de colesterol, la mayoría de personas tienen un nivel de colesterol entre los 200 y los 300.

```python
sns.displot(df["chol"], kde=True)
plt.show()
```

Los pacientes con angina inducida por el ejercicio (exang=1) tienden a tener una menor frecuencia cardíaca máxima (thalach), lo que sugiere una posible limitación en su capacidad cardiovascular.

```python
sns.boxplot(data=df, x="exang", y="thalach")
plt.show()
```

Muestra las personas de determinada edad que tienen depresión.

```python
sns.scatterplot(data=df, x="age", y="oldpeak")
plt.show()
```

Este método permite resumir los datos obteniendo el promedio de cada variable numérica según la edad, lo que es útil para análisis estadísticos y visualización de tendencias y tener en cuenta la información dada.

```python
df.groupby("age").mean().reset_index()
```

Este método permite encontrar el nivel máximo de colesterol en cada grupo de sexo, lo que es útil para análisis comparativos entre poblaciones.

```python
df.groupby("sex")["chol"].max().reset_index()
```
